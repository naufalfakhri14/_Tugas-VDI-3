```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Mmembaca dataset
df = pd.read_csv('marketing_campaign.csv', sep='\t')

# Menghitung usia dari tahun lahir
current_year = 2024 #dalam kurun waktu 2024
df['Age'] = current_year - df['Year_Birth']  #inisialisasikan

# Membuat visualisasi pengeluaran berdasarkan pendapatan dan usia
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Income', y='MntWines', hue='Age', palette='viridis', alpha=0.7)  #membuat variabel x,y,z

# membuat judul dan label
plt.title('Transaksi anggur berdasarkan pendapatan dan usia pelanggan')
plt.xlabel('Pendapatan pertahun')
plt.ylabel('Pengeluaran  Anggur')
plt.grid()
plt.legend(title='Usia', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Membuat pie chart proporsi pengeluaran
labels = ['Wines', 'Fruits', 'Meat Products', 'Fish Products', 'Sweet Products', 'Gold Prods']
sizes = [df['MntWines'].sum(), df['MntFruits'].sum(), df['MntMeatProducts'].sum(), df['MntFishProducts'].sum(), df['MntSweetProducts'].sum(), df['MntGoldProds'].sum()]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart Proporsi Pengeluaran')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('marketing_campaign.csv', sep='\t')

# Membuat histogram pengeluaran untuk anggur
plt.hist(df['MntWines'], bins=10, edgecolor='red')
plt.title('Histogram transaksi Anggur')
plt.xlabel('Pengeluaran')
plt.ylabel('Frekuensi')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Membuat line chart tren pengeluaran
years = df['Year_Birth'].apply(lambda x: 2024 - x)
wines_expenses = df['MntWines']

plt.figure(figsize=(10, 6))
plt.plot(years, wines_expenses, marker='o')
plt.title('Line Chart Tren Pengeluaran Anggur')
plt.xlabel('Tahun')
plt.ylabel('Pengeluaran Anggur')
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca dataset
# Use the sep='\t' argument to specify that the separator is a tab
df = pd.read_csv('marketing_campaign.csv', sep='\t')

# Menghitung mean, median, mode pengeluaran anggur
mean_wines = df['MntWines'].mean()
median_wines = df['MntWines'].median()
mode_wines = df['MntWines'].mode()[0]

# Menyiapkan data untuk visualisasi
labels = ['Mean', 'Median', 'Modus']
values = [mean_wines, median_wines, mode_wines]

# Membuat bar chart
plt.figure(figsize=(8, 5))
plt.bar(labels, values, color=['red', 'yellow', 'blue'])
plt.title('Statistik transaksi pengeluaran Anggur')
plt.ylabel('Pengeluaran (dalam unit mata uang)')
plt.xlabel('Statistik')
plt.grid(axis='y', alpha=0.75)

# Menambahkan nilai di atas bar
for i in range(len(values)):
    plt.text(i, values[i], f'{values[i]:.2f}', ha='center', va='bottom')

plt.show()
```
